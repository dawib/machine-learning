# Zadanie 1

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2$, startując z punktu $x=2$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def f(x):
    return x**2

def grad(x):
    return 2*x

x = 2
a = .1
i = 0

while np.abs(-2*x)>0.00001:
    d = -grad(x)
    while f(x+a*d)>f(x):
        a/=2
    x1 = x+a*d
    if i<5:
        plt.arrow(x,f(x), x1-x,f(x1)-f(x), color='red', width=0.03, head_width=0.15)
    i+= 1
    x = x1
    
Xf = np.linspace(-3,3,100)
Yf = f(Xf)

plt.plot(Xf,Yf, '--')
plt.show()
print(x)
```

# Zadanie 2

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x,y)=x^2-2x+y^2$, startując z punktu $(x,y)=(2,2)$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

```python
def f(p):
    x,y = p
    return x**2-2*x+y**2

def grad(p):
    x,y = p
    gx = 2*x-2
    gy = 2*y
    return np.array([gx,gy])

p0 = np.array([2,2])
x = [p0[0]]
y = [p0[1]]
a = .2

i = 0
while np.sum(grad(p0))>0.00000001 and i<10000:
    d = -grad(p0)
    while f(p0+a*d)>f(p0):
        a/=2
    p0 = p0+a*d
    if i<5:
        x.append(p0[0])
        y.append(p0[1])
    i+=1
    
print(p0)

X,Y = np.mgrid[-2:4:100j,-3:3:100j]
Z = X**2 - 2*X + Y**2
plt.contourf(X,Y,Z, levels=range(-3,6), cmap=plt.cm.coolwarm, alpha=0.5)
plt.plot(x,y,color='red')
plt.show()
```

# Zadanie 4
Zapoznaj się z funkcją <tt>optimize.fmin_cg</tt>
https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin_cg.html
i znajdż minum funkcji

$$
f(x,y) = (x+1)^2+y^2
$$

```python
import numpy as np
from scipy import optimize

def f(x):
    u,v = x
    return (u+1)**2+v**2

def grad(x):
    u,v = x
    gu = 2*(u+1)
    gv = 2*v
    return np.asarray((gu,gv))

x0 = np.asarray((10,-6))
result = optimize.fmin_cg(f, x0)
print(result)

i = 0
a=.2
x0 = np.asarray((10,-6))
while i<100000 and np.dot(grad(x0),grad(x0))>0.00000000001:
    d = -grad(x0)
    while f(x0+a*d)>f(x0):
        a/=2
    x0 = x0+a*d
    i+=1
    a=.2
    
print(x0)
```

# Zadanie 5
Zdefiniuj funkcję opisującą błąd kwadratowy:

$$
\mathrm{compute\_error}(a,b) =  \sum_{i=1}^{n} ( y_i - (ax_i+b) )^2.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm

f = lambda x: (x**2)


x = np.array([.2, .5, .8, .9, 1.3, 1.7, 2.1, 2.7, 7.7])
y = f(x) + np.random.randn(len(x))
```

```python
#...
def compute_error(P):
    a,b = P
    return np.sum((y-(a*x+b))**2)

```

```python
print(compute_error(np.asarray((1,1))))
print(compute_error(np.array([1,0])))
```

```python
x_tr = np.linspace(0., 8, 200)
y_tr = f(x_tr)

# model.
lr = lm.LinearRegression()
lr.fit(x[:, np.newaxis], y);
print(lr.coef_)
print(lr.intercept_)

f_lr = lambda x: lr.coef_ * x +lr.intercept_

x_f_lr = np.linspace(0., 8, 200)
y_f_lr = f_lr(x_tr)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k');
plt.plot(x_f_lr, y_f_lr, 'g');
plt.plot(x, y, 'ok', ms=10);
plt.show()
```

```python


def grad(P):
    a,b = P
    h = 0.000000001
    ga = (compute_error(np.asarray((a+h,b)))-compute_error(np.asarray((a,b))))/h
    gb = (compute_error(np.asarray((a,b+h)))-compute_error(np.asarray((a,b))))/h
    return np.asarray((ga,gb))

P0 = np.asarray((1,0))

i = 0
a = .2
while i<5000 and np.dot(grad(P0),grad(P0))>0.0000000001:
    d = -grad(P0)
    while compute_error(P0+a*d) > compute_error(P0):
        a/=2
    P0 = P0+a*d
    i+=1
    a=.2
    #print(i)
#print(i)
    
print(P0)
print(compute_error(P0))

x_f_lr = np.linspace(0., 8, 200)
y_f_lr = f_lr(x_tr)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k');
plt.plot(x_f_lr, y_f_lr, 'g');
y_pred = P0[0]*x_tr + P0[1]
plt.plot(x_f_lr, y_pred+0.1, 'r');
plt.plot(x, y, 'ok', ms=10);
plt.show()
```
